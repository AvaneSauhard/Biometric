```python
# function created to perform upsampling
# outer loop selects a sample for which the new samples are to be generated
# inner loops are used to generate random normal noise using mean 0 and standard deviation 1
# this noise is then added to the original samples
def generatenoise(data_n):
  dfn  = pd.DataFrame()
  
  for row in range(len(data_n)):
    mu = 0
    sigma = 1
    for j in range(0,5):
      noise = np.random.normal(mu,sigma,size = data_n.iloc[[row]].shape)
      df  = pd.DataFrame()
      df = data_n.iloc[[row]] + 0.2*noise
      dfn = pd.concat([dfn,df], axis = 0, ignore_index = True)

    for j in range(0,5):
      noise = np.random.normal(mu,sigma,size = data_n.iloc[[row]].shape)
      df  = pd.DataFrame()
      df = data_n.iloc[[row]] + 0.3*noise
      dfn = pd.concat([dfn,df], axis = 0, ignore_index = True)
      
  return dfn

```

```python
def balance(data_n, c):
  dfn  = pd.DataFrame()
  
  for row in range(0,c):
    mu = 0
    sigma = 1
   
    noise = np.random.normal(mu,sigma,size = data_n.iloc[[row]].shape)
    df  = pd.DataFrame()
    df = data_n.iloc[[row]] + 0.2*noise
    dfn = pd.concat([dfn,df], axis = 0, ignore_index = True)
  return dfn
```

```python
def build_model():
  model = tf.keras.models.Sequential()
  #  layers are added to build a cnn model
  model.add(tf.keras.layers.Conv1D(filters = 64, kernel_size = 8, activation = 'relu',  input_shape = (800,4)))
  model.add(tf.keras.layers.BatchNormalization())
  model.add(tf.keras.layers.MaxPooling1D(pool_size = (3), strides = (1)))
  
  model.add(tf.keras.layers.Conv1D(filters = 128, kernel_size = 8, activation = 'relu', input_shape = (791,64)))
  model.add(tf.keras.layers.BatchNormalization())
  model.add(tf.keras.layers.MaxPooling1D(pool_size = (3), strides = (1)))

  model.add(tf.keras.layers.Conv1D(filters = 256, kernel_size = 8, activation = 'relu',  input_shape = (782,128)))
  model.add(tf.keras.layers.BatchNormalization())
  model.add(tf.keras.layers.MaxPooling1D(pool_size = (3), strides = (1)))

  model.add(tf.keras.layers.Flatten())

  model.add(tf.keras.layers.Dense(units = 64, activation= 'relu'))

  model.add(tf.keras.layers.Dense(units = 64, activation= 'relu'))
  
  model.add(tf.keras.layers.Dense(units = 20, activation= 'softmax'))
 
  model.compile(optimizer = 'Adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
  return model
model = build_model()
model.summary()

```
